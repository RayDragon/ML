```python
n=256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.cos(2*X)

pt.plot(X, Y, color='blue', alpha=1)
pt.fill_between(X, 1, Y + 1, color='green', alpha=0.25)
pt.plot(X,Y-1, color='purple')
pt.fill_between(X, -1, Y -1, (Y-1)>-1, color='pink', alpha=.5)
pt.fill_between(X, -1, Y -1, (Y-1)<-1, color='yellow', alpha=.6)
pt.show()
```
